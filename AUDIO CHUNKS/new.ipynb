```python
import os
from pydub import AudioSegment
from pydub.silence import split_on_silence
from warnings import warn  
```

```python
name_list = os.listdir('Z:/audios/T_1') 
try:
    count = 0
    for name in name_list:
        count = count +1
        sound_file = AudioSegment.from_wav("Z:/audios/T_1/"+str(name))
        audio_chunks = split_on_silence(sound_file, min_silence_len=1900, silence_thresh=sound_file.dBFS,  keep_silence=300 )
        name = name.replace('.wav','')
        for i, chunk in enumerate(audio_chunks):
            out_file = "mysorteddata/chunk/"+ str(count)+name+"-{0}.wav".format(i+1)
            print("exporting", out_file)
            chunk.export(out_file, format="wav")
except (FileNotFoundError, Warning, warn, RuntimeWarning):
    pass
```
